```python
import matplotlib.pyplot as plt

# Sample data
reactions = ['Fever', 'Fatigue', 'Headache', 'Chills', 'Vomiting', 'Diarrhea', 'Myalgia', 'Arthralgia', 'Injection site pain']

mild_HD = [12, 22, 12, 2, 6, 12, 6, 12, 38]
moderate_HD = [8, 14, 10, 4, 6, 10, 6, 6, 20]
severe_HD = [0, 0, 0, 0, 0, 0, 0, 0, 0]

mild_HCW = [24, 12, 14, 6, 6, 12, 6, 12, 50]
moderate_HCW = [14, 10, 14, 6, 6, 10, 6, 6, 18]
severe_HCW = [0, 0, 0, 0, 0, 0, 0, 0, 0]

# Plotting
fig, ax = plt.subplots(figsize=(10,7))

# Stacking bars for HD
ax.barh(reactions, mild_HD, label='Mild HD', color='green')
ax.barh(reactions, moderate_HD, left=mild_HD, label='Moderate HD', color='blue')
ax.barh(reactions, severe_HD, left=[i+j for i,j in zip(mild_HD, moderate_HD)], label='Severe HD', color='red')

# Stacking bars for HCW
ax.barh(reactions, mild_HCW, left=[i+j+k for i,j,k in zip(mild_HD, moderate_HD, severe_HD)], label='Mild HCW', color='lightgreen')
ax.barh(reactions, moderate_HCW, left=[i+j+k+l for i,j,k,l in zip(mild_HD, moderate_HD, severe_HD, mild_HCW)], label='Moderate HCW', color='lightblue')
ax.barh(reactions, severe_HCW, left=[i+j+k+l+m for i,j,k,l,m in zip(mild_HD, moderate_HD, severe_HD, mild_HCW, moderate_HCW)], label='Severe HCW', color='pink')

# Setting labels and title
ax.set_xlabel('Percentage (%)')
ax.set_title('Adverse Reactions to mRNA Vaccine by Group')
ax.legend()

plt.tight_layout()
plt.show()
```
